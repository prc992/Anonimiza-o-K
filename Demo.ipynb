```python
import pandas as pd
```

```python
df = pd.read_csv("dados/teste2.csv", encoding = 'utf-8')
```

```python
df
```

```python
listaCamposSemiID = ["dataNascimento", "municipioCaso", "racaCor","sexoCaso"]
```

```python
listaCamposSens = ["comob1","comob2","resExame"]
```

```python
listaCamposSaida = listaCamposSemiID + listaCamposSens
```

```python
strBlank = len(listaCamposSemiID)*'*'
```

```python
todasColunas = df.columns.to_list()
```

```python
listaCamposRemovidas = [item for item in todasColunas if item not in listaCamposSaida]
```

```python
def criaIdentificador(row,ListaCampos):
    strID = ""
    for x in ListaCampos:
        strID = strID + str(row [x])
    return strID
```

```python
def criaContador(df,campo):
    dfInterno = df[campo].value_counts().to_frame('Contador')
    dfInterno[campo] = dfInterno.index
    strQtd = 'qtd-' + campo

    try:
        df = df.drop(columns=[strQtd])
    except:
        pass
    
    dfInterno.columns = [strQtd,campo]
    return df.merge(dfInterno, left_on=campo, right_on=campo)
```

```python
%%time
df['id'] = df.apply(criaIdentificador, axis=1,args=(listaCamposSemiID,))
```

```python
df
```

```python
%%time
df = criaContador(df,'id')
```

```python
df
```

```python
def retornaKAnom (df,strBlank):
    df1 = df[df["id"]!=strBlank]
    dfInterno = df1['qtd-id'].value_counts().to_frame('Contador')
    dfInterno['qtd-id'] = dfInterno.index
    return min(dfInterno['qtd-id'])
```

```python
retornaKAnom(df,strBlank)
```

```python
def suprimeCampo(row,nomeCampo,k):
    retorno = row[nomeCampo]
    if row['qtd-id'] < k:
        retorno = '*'
        
    return retorno
```

```python
k=2
```

```python
df["dataNascimento"] = df.apply(suprimeCampo, axis=1,args=("dataNascimento",k,))
```

```python
df
```

```python
%%time
df['id'] = df.apply(criaIdentificador, axis=1,args=(listaCamposSemiID,))
df = criaContador(df,'id')
retornaKAnom(df,strBlank)
```

```python
df
```

```python
df["racaCor"] = df.apply(suprimeCampo, axis=1,args=("racaCor",k,))
```

```python
%%time
df['id'] = df.apply(criaIdentificador, axis=1,args=(listaCamposSemiID,))
df = criaContador(df,'id')
retornaKAnom(df,strBlank)
```

```python
%%time
df["municipioCaso"] = df.apply(suprimeCampo, axis=1,args=("municipioCaso",k,))
df['id'] = df.apply(criaIdentificador, axis=1,args=(listaCamposSemiID,))
df = criaContador(df,'id')
retornaKAnom(df,strBlank)
```

```python
%%time
df["sexoCaso"] = df.apply(suprimeCampo, axis=1,args=("sexoCaso",k,))
df['id'] = df.apply(criaIdentificador, axis=1,args=(listaCamposSemiID,))
df = criaContador(df,'id')
```
